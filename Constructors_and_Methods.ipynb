## Summary: Key Takeaways

| Feature | Description | Example |
|---------|-------------|---------|
| **Constructor (`__init__`)** | Initializes instance attributes when object is created | `def __init__(self, name):` |
| **Instance Method** | Operates on instance attributes; first parameter is `self` | `def get_info(self):` |
| **Class Method** | Works with class attributes; uses `@classmethod` decorator; first parameter is `cls` | `@classmethod` with `cls` |
| **Static Method** | Utility function in class; no `self` or `cls`; uses `@staticmethod` | `@staticmethod def is_valid():` |
| **Method Overriding** | Child class provides different implementation of parent method | `class Dog(Animal): def speak():` |
| **super()** | Calls parent class method | `super().__init__(name)` |
| **Polymorphism** | Different objects respond to same method call differently | Same `speak()` method, different output |

### When to Use Each Method Type:

✅ **Instance Methods**: Most common - operate on instance data
✅ **Class Methods**: Creating alternative constructors, managing class-level data
✅ **Static Methods**: Utility functions logically related to the class but independent of instance/class state
✅ **Constructors**: Always needed - initialize all required instance attributes

### Practice Tips:
1. Always use `self` in instance methods
2. Use `cls` in class methods with `@classmethod` decorator
3. Use `@staticmethod` for utility functions
4. Create constructors to ensure proper initialization
5. Use inheritance and method overriding for code reuse and polymorphism

```python
class BankAccount:
    """A class to simulate a bank account"""
    
    # Class attribute - shared by all instances
    bank_name = "Python Bank"
    total_accounts = 0
    
    def __init__(self, account_holder, initial_balance):
        """Constructor - initializes a bank account"""
        self.account_holder = account_holder
        self.balance = initial_balance
        self.transactions = []
        self.transactions.append(f"Account opened with ${initial_balance}")
        BankAccount.total_accounts += 1
    
    def deposit(self, amount):
        """Add money to the account"""
        self.balance += amount
        self.transactions.append(f"Deposited: ${amount}")
        return f"✓ Deposited ${amount}. New balance: ${self.balance}"
    
    def withdraw(self, amount):
        """Remove money from the account"""
        if amount > self.balance:
            return f"✗ Insufficient funds! Balance: ${self.balance}"
        self.balance -= amount
        self.transactions.append(f"Withdrew: ${amount}")
        return f"✓ Withdrew ${amount}. New balance: ${self.balance}"
    
    def get_balance(self):
        """Returns current balance"""
        return self.balance
    
    def get_statement(self):
        """Returns account statement"""
        statement = f"\n{'='*50}\n"
        statement += f"{BankAccount.bank_name} - Account Statement\n"
        statement += f"Account Holder: {self.account_holder}\n"
        statement += f"Current Balance: ${self.balance}\n"
        statement += f"\nTransactions:\n"
        for i, transaction in enumerate(self.transactions, 1):
            statement += f"  {i}. {transaction}\n"
        statement += f"{'='*50}\n"
        return statement

# Create multiple independent bank accounts
print("=== Creating Multiple Bank Accounts ===\n")

account1 = BankAccount("Alice Johnson", 1000)
account2 = BankAccount("Bob Smith", 500)
account3 = BankAccount("Carol White", 2000)

print(f"Total accounts created: {BankAccount.total_accounts}")

# Each account maintains independent state
print("\n=== Performing Operations on Different Accounts ===\n")

print("Account 1:")
print(account1.deposit(500))
print(account1.withdraw(200))

print("\nAccount 2:")
print(account2.deposit(300))
print(account2.withdraw(100))

print("\nAccount 3:")
print(account3.withdraw(500))
print(account3.deposit(1000))

# Show that each account maintains its own state
print("\n=== Account Statements (Independent State) ===")
print(account1.get_statement())
print(account2.get_statement())
print(account3.get_statement())

# Summary
print("=== Summary ===")
print(f"Alice's balance: ${account1.get_balance()}")
print(f"Bob's balance: ${account2.get_balance()}")
print(f"Carol's balance: ${account3.get_balance()}")
print(f"Total accounts in {BankAccount.bank_name}: {BankAccount.total_accounts}")
```

## 7. Practice with Multiple Instances

### Key Concept: Instance Independence
Each instance of a class:
- Has its own copy of instance attributes
- Calls the same methods defined in the class
- Maintains independent state from other instances

### Example: Bank Account Management

```python
# Parent class
class Animal:
    """Parent class representing an animal"""
    
    def __init__(self, name):
        """Constructor for Animal"""
        self.name = name
    
    def speak(self):
        """Default speak method - will be overridden"""
        return f"{self.name} makes a sound"
    
    def move(self):
        """Default move method"""
        return f"{self.name} is moving"

# Child class 1: Dog (overrides speak method)
class Dog(Animal):
    """Dog class that inherits from Animal"""
    
    def __init__(self, name, breed):
        """Constructor for Dog - calls parent constructor using super()"""
        super().__init__(name)  # Call parent's __init__
        self.breed = breed
    
    # Override speak method
    def speak(self):
        """Dogs bark - overrides parent's speak method"""
        return f"{self.name} (Breed: {self.breed}) says: Woof! Woof! 🐕"

# Child class 2: Cat (overrides speak method)
class Cat(Animal):
    """Cat class that inherits from Animal"""
    
    def __init__(self, name, color):
        """Constructor for Cat"""
        super().__init__(name)  # Call parent's __init__
        self.color = color
    
    # Override speak method
    def speak(self):
        """Cats meow - overrides parent's speak method"""
        return f"{self.name} (Color: {self.color}) says: Meow! Meow! 🐱"

# Child class 3: Bird (overrides move method)
class Bird(Animal):
    """Bird class that inherits from Animal"""
    
    def speak(self):
        """Birds chirp"""
        return f"{self.name} says: Tweet! Tweet! 🐦"
    
    def move(self):
        """Birds fly - overrides parent's move method"""
        return f"{self.name} is flying gracefully in the sky ✈️"

print("=== Demonstrating Method Overriding and Polymorphism ===\n")

# Create different animals
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers", "Orange")
bird = Bird("Tweety")

# Same method call, different behavior (Polymorphism!)
animals = [dog, cat, bird]

print("1. Calling speak() on different animals:")
for animal in animals:
    print(f"   {animal.speak()}")

print("\n2. Calling move() on different animals:")
for animal in animals:
    print(f"   {animal.move()}")

# Access parent's method using super() example
print("\n3. Accessing parent's method (without override):")
print(f"   {dog.move()}")  # Uses parent's move method (not overridden)
```

## 6. Demonstrate Method Overriding

### What is Method Overriding?
**Method overriding** occurs when a child class provides a different implementation of a method that exists in the parent class.
- Used for **polymorphism** - different objects can respond to the same method call differently
- The child class's method replaces the parent class's method
- Call `super()` to access the parent's method if needed

### Example: Parent and Child Classes

```python
class Calculator:
    """A class with utility calculation methods"""
    
    def __init__(self, name):
        """Constructor - stores calculator name"""
        self.name = name
    
    # Instance method
    def describe(self):
        """Returns description of this calculator"""
        return f"This is {self.name} calculator"
    
    # Static method 1: Add two numbers
    @staticmethod
    def add(a, b):
        """Adds two numbers (doesn't use self or cls)"""
        return a + b
    
    # Static method 2: Subtract two numbers
    @staticmethod
    def subtract(a, b):
        """Subtracts two numbers"""
        return a - b
    
    # Static method 3: Check if valid percentage
    @staticmethod
    def is_valid_percentage(value):
        """Checks if a value is a valid percentage (0-100)"""
        return isinstance(value, (int, float)) and 0 <= value <= 100
    
    # Static method 4: Convert temperature
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        """Converts Celsius to Fahrenheit"""
        return (celsius * 9/5) + 32

# Create calculator instance
calc = Calculator("Scientific")

print("=== Instance Method ===")
print(calc.describe())

print("\n=== Static Methods ===")

# Call static methods from the instance
print(f"\n1. Using static methods through instance:")
print(f"   {calc.add(10, 5)} = 15")
print(f"   {calc.subtract(10, 5)} = 5")

# Call static methods from the class (more common)
print(f"\n2. Using static methods through class:")
print(f"   Calculator.add(20, 8) = {Calculator.add(20, 8)}")
print(f"   Calculator.subtract(20, 8) = {Calculator.subtract(20, 8)}")

# Static utility methods
print(f"\n3. Validation with static method:")
print(f"   Is 75 a valid percentage? {Calculator.is_valid_percentage(75)}")
print(f"   Is 150 a valid percentage? {Calculator.is_valid_percentage(150)}")

print(f"\n4. Conversion with static method:")
print(f"   25°C = {Calculator.celsius_to_fahrenheit(25)}°F")
print(f"   0°C = {Calculator.celsius_to_fahrenheit(0)}°F")
```

## 5. Create Static Methods

### What are Static Methods?
**Static methods** are methods that don't access instance or class attributes.
- Use the `@staticmethod` decorator
- No `self` or `cls` parameter
- Act like regular functions but belong to the class namespace
- Useful for utility functions related to the class

### Example: Using Static Methods

```python
class Student:
    """Class to represent a student with class methods"""
    
    # Class attribute (shared by all instances)
    total_students = 0
    school_name = "Python Academy"
    
    def __init__(self, name, student_id):
        """Constructor - creates a student instance"""
        self.name = name
        self.student_id = student_id
        Student.total_students += 1  # Increment class attribute
    
    # Instance method
    def get_info(self):
        """Returns student information"""
        return f"Student: {self.name} (ID: {self.student_id})"
    
    # Class method 1: Alternative constructor
    @classmethod
    def from_string(cls, student_string):
        """Create a student from a formatted string like 'Alice-101'"""
        name, student_id = student_string.split('-')
        return cls(name, student_id)
    
    # Class method 2: Get total students
    @classmethod
    def get_total_students(cls):
        """Returns the total number of students created"""
        return f"Total students in {cls.school_name}: {cls.total_students}"
    
    # Class method 3: Change school name
    @classmethod
    def change_school(cls, new_school):
        """Changes the school name for all students"""
        cls.school_name = new_school
        print(f"✓ School changed to {new_school}")

# Create students using the regular constructor
student1 = Student("Alice", "101")
student2 = Student("Bob", "102")

print("=== Instance Method ===")
print(student1.get_info())
print(student2.get_info())

print("\n=== Class Methods ===")

# Call class method: get_total_students
print(f"1. {Student.get_total_students()}")

# Call class method: alternative constructor
print(f"\n2. Creating student from string using from_string():")
student3 = Student.from_string("Charlie-103")
print(student3.get_info())
print(Student.get_total_students())

# Call class method: change_school
print(f"\n3. Change school name:")
Student.change_school("Python University")

# Show class attribute is shared
print(f"\nAccessing class attribute from instance: {student1.school_name}")
print(f"Accessing class attribute from class: {Student.school_name}")
```

## 4. Create Class Methods

### What are Class Methods?
**Class methods** are methods that work with class attributes (shared by all instances) rather than instance attributes.
- Use the `@classmethod` decorator
- First parameter is `cls` (not `self`), which refers to the class itself
- Useful for creating alternative constructors or modifying class-level data

### Example: Using Class Methods

```python
class Person:
    """A class to represent a person with instance methods"""
    
    def __init__(self, name, age, email):
        """Constructor - initializes a Person object"""
        self.name = name
        self.age = age
        self.email = email
    
    # Instance method 1: Get info
    def get_info(self):
        """Returns a string with person's information"""
        return f"{self.name} ({self.age} years old) - {self.email}"
    
    # Instance method 2: Update age
    def have_birthday(self):
        """Increments age by 1"""
        self.age += 1
        return f"🎉 Happy Birthday! {self.name} is now {self.age} years old"
    
    # Instance method 3: Check if adult
    def is_adult(self):
        """Checks if person is 18 or older"""
        return self.age >= 18

# Create instances and call instance methods
alice = Person("Alice", 25, "alice@email.com")
bob = Person("Bob", 17, "bob@email.com")

print("=== Calling Instance Methods ===")
print(f"\n1. get_info() method:")
print(f"   Alice: {alice.get_info()}")
print(f"   Bob: {bob.get_info()}")

print(f"\n2. have_birthday() method:")
print(f"   {bob.have_birthday()}")  # Bob becomes 18
print(f"   Updated info: {bob.get_info()}")

print(f"\n3. is_adult() method:")
print(f"   Alice is adult: {alice.is_adult()}")
print(f"   Bob is adult: {bob.is_adult()}")
```

## 3. Create Instance Methods

### What are Instance Methods?
**Instance methods** are functions defined inside a class that operate on instance attributes.
- First parameter is always `self`
- Can access and modify instance attributes
- Called on objects created from the class

### Example: Adding Methods to Person Class

```python
class Person:
    """A class to represent a person"""
    
    def __init__(self, name, age, email):
        """
        Constructor - initializes a Person object
        
        Parameters:
        - self: refers to the instance being created
        - name: person's name
        - age: person's age
        - email: person's email address
        """
        self.name = name      # Instance attribute
        self.age = age        # Instance attribute
        self.email = email    # Instance attribute
        print(f"✓ Person object created for {name}")

# Create instances (objects) of the Person class
person1 = Person("Alice", 25, "alice@email.com")
person2 = Person("Bob", 30, "bob@email.com")

# Access instance attributes
print(f"\nPerson 1: {person1.name}, Age: {person1.age}")
print(f"Person 2: {person2.name}, Age: {person2.age}")
```
