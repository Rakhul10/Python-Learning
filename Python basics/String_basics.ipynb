```python
# find
txt = "Thank you for the music\nWelcome to the jungle"
print(txt.find("x",0,20)) #returns the lowest index of the substring if it is found in the string. If it is not found, it returns -1. 

# index
txt = "Thank you for the music\nWelcome to the jungle"
print(txt.index("x",0,20)) #returns the lowest index of the substring if it is found in the string. If it is not found, it raises a ValueError.
```

```python
num="789"
print(num.rjust(10,"#")) #returns a right justified string of a given minimum width. The original string is returned if the specified width is less than or equal to the length of the original string. The fillchar is used as the padding character. It defaults to a space.
print(num.ljust(10,"#")) #returns a left justified string of a given minimum width. The original string is returned if the specified width is less than or equal to the length of the original string. The fillchar is used as the padding character. It defaults to a space.
print(num.center(10,"#")) #returns a centered string of a given minimum width. The original string is returned if the specified width is less than or equal to the length of the original string. The fillchar is used as the padding character. It defaults to a space.
print(num.zfill(10)) #returns a copy of the string with '0' characters padded to the left of the original string until the total width is equal to the specified width. The original string is returned if the specified width is less than or equal to the length of the original string. If the original string already starts with a sign ('+' or '-'), the '0' characters are inserted after the sign.
```

```python
num="345"
print(num.encode()) #returns an encoded version of the string as a bytes object. The default encoding is 'utf-8'. The errors parameter specifies how to handle encoding errors. It defaults to 'strict', which raises a UnicodeEncodeError on encoding errors. Other possible values are 'ignore', 'replace', 'xmlcharrefreplace', 'backslashreplace', and any other name registered via codecs.register_error().
print(num.encode("ascii", "ignore")) 
print(type(num.encode()))  
```

```python
#split() - splits a string into a list. The splitting is done at the specified separator. The default separator is any whitespace. The maxsplit parameter specifies the maximum number of splits to do. The default value is -1, which means "all occurrences".
txt = "Hello welcome to my world." 
x = txt.split(" ")
print(x)

for i in txt.split(" "):
    print(i)
```

```python
#join
txt = ['Hello', 'welcome', 'to', 'my', 'world.']
print("_".join(txt)) #returns a string which is the concatenation of the strings in the iterable. The separator between the elements is the string on which the method is called.

txt = ['Hello', 'welcome', 'to', 'my', 'world', 1,2,3]
print(" ".join(str(i) for i in txt)) #returns a string which is the concatenation of the strings in the iterable.
```

```python
txt="Hello welcome to 45 my world 1 2 3"
for i in txt:
    if i.isdigit():
        print(i)

for i in txt.split(" "):
    if i.isalpha():
        print(i)
# to extact only the words from the string
char=""
for i in txt.split(" "):
    if i.isalpha():
        char += i + " "
print(char)
```

```python
# if else in Fstrings
price = 49
txt = f"It is very {'Expensive' if price>50 else 'Cheap'}"

print(txt)

#math functions in strings
price = 59
tax = 0.25
txt = f"The price is {price + (price * tax)} dollars"
print(txt)

#decimal formatting in Fstrings
price = 59
txt = f"The price is ${price:.2f} dollars" #formats the price to 2 decimal places
print(txt)

#excute a function in Fstrings
def myconverter(x):
  return x * 0.3048

txt = f"The plane is flying at a {myconverter(30000)} meter altitude"
print(txt)
```
