A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

syntax-> lambda arguments : expression

It can only contain expressions and can’t include statements in its body. ->  statements like return, pass, assert, or raise will raise a SyntaxError exception
It is written as a single line of execution.
It does not support type annotations.
It can be immediately invoked (IIFE).

#lamda arguments are evaluated at runtime, not when the lambda is defined. This means that if you use a variable in a lambda, it will use the value of that variable at the time the lambda is called, not when it was defined.

```python
x = lambda a : a + 10
print(x(5))
```

```python
(lambda x: (x % 2 and 'odd' or 'even'))(3)
```

```python
#multiple arguments
x = lambda a, b : a * b
print(x(5, 6))
```

Why Use Lambda Functions?
The power of lambda is better shown when you use them as an anonymous function inside another function.

Say you have a function definition that takes one argument, and that argument will be multiplied with an unknown number:

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))  #this part is tricky
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

```python
# higher-order function (a function that takes another function as a parameter)
high_ord_func = lambda x, func: x + func(x)
high_ord_func(2, lambda x: x * x)
```

```python
def full_name(first: str, last: str) -> str:
    return f'{first.title()} {last.title()}'
print(full_name('john', 'doe'))

#Lambda syntax doesn't support type annotations
#If you need type hints, use a def function instead
lambda first: str, last: str: first.title() + " " + last.title() -> str
```

Alternative to Lambda -> Map, Filter, reduce

```python
#The map() function applies a specific function to every item in an iterable (like a list) and returns an iterator.
numbers = [1, 2, 3, 4]
# Double every number
squared = list(map(lambda x: x * 2, numbers))
print(squared)  # Output: [2, 4, 6, 8]
```

```python
#The filter() function creates a new iterator from elements of an iterable for which a function returns True.
numbers = [1, 2, 3, 4, 5, 6]
# Keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

```python
#Unlike the first two, reduce() is not a built-in function in modern Python; it lives in the functools module. It applies a rolling computation to sequential pairs of values until only one single value remains.
from functools import reduce

numbers = [1, 2, 3, 4]
# Calculate the product of all numbers (1*2*3*4)
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24

# Find maximum
reduce(lambda x, y: x if x > y else y, [5, 2, 8, 1])  # Output: 8
```
