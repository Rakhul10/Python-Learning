#iteration in class
#When you write a class, you can make its objects behave like lists or other collections — meaning you can loop over them with for


```python
class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end
        self.current = start

    def __iter__(self):   # returns iterator object
        return self

    def __next__(self):   # defines next value
        if self.current >= self.end:
            raise StopIteration   # signals end of loop
        value = self.current
        self.current += 1
        return value

nums = MyRange(1, 5)
#Python calls nums.__iter__() → returns self (the object itself)
for n in nums: #The for loop automatically calls __next__() repeatedly until StopIteration is raised!
    print(n)

```

```python
#__getitem__
class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __getitem__(self, index):
        if index < 0 or self.start + index >= self.end:
            raise IndexError("Index out of range")
        #Python automatically increments the index for each iteration of the loop, starting from 0
        return self.start + index
nums = MyRange(1, 5)
for n in nums: #The for loop automatically calls __getitem__() with increasing index until IndexError is raised!
    print(n)
```
