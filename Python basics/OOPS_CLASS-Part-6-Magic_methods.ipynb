Magic Methods—more commonly called Dunder Methods (short for Double Underscore)—are built-in methods that begin and end with two underscores. Examples include __init__, __str__, and __add__.

They are called "magic" because you rarely call them directly. Instead, Python triggers them automatically behind the scenes when you use standard operators or built-in functions on your custom objects.

By defining these same double-underscore methods in your custom classes, you give them the "magic" ability to hook into Python's built-in operators and function

Operator,Dunder Method,Trigger Condition
+,"__add__(self, other)",obj1 + obj2
-,"__sub__(self, other)",obj1 - obj2
*,"__mul__(self, other)",obj1 * obj2

Operator,Dunder Method,Trigger Condition
==,"__eq__(self, other)",obj1 == obj2
<,"__lt__(self, other)",obj1 < obj2
>,"__gt__(self, other)",obj1 > obj2

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # 1. Customizing how it prints
    def __str__(self):
        return f"{self.name} costs ₹{self.price}"

    # 2. Customizing the '+' operator
    def __add__(self, other):
        # Returns the total cost of both products combined
        return self.price + other.price

    # 3. Customizing the '==' operator
    def __eq__(self, other):
        # Products are equal if they have the same name and price
        return self.name == other.name and self.price == other.price

p1 = Product("Shirt", 1500)
p2 = Product("Jeans", 2500)
p3 = Product("Shirt", 1500)

# Triggers __str__
print(p1)  
# Output: Shirt costs ₹1500

# Triggers __add__
total_bill = p1 + p2  
print(total_bill)  
# Output: 4000

# Triggers __eq__
print(p1 == p3)  
# Output: True

 
```

```python
# Making Your Object Work Like a Collection/List
class Playlist:
    def __init__(self, songs):
        self.songs = songs  # Expects a list of song names

    def __len__(self):
        return len(self.songs)

    def __getitem__(self, index):
        return self.songs[index]

my_playlist = Playlist(["Song A", "Song B", "Song C"])

print(len(my_playlist))      # Output: 3 (Triggers __len__)
print(my_playlist[1])        # Output: Song B (Triggers __getitem__)
```

```python
#Making Your Object Callable Like a Function
class Greeter:
    def __init__(self, greeting):
        self.greeting = greeting

    def __call__(self, name): # This method allows the object to be called like a function
        return f"{self.greeting}, {name}!"

welcome = Greeter("Vanakkam")
# Calling the object directly like a function:
print(welcome("Anand"))  # Output: Vanakkam, Anand!
```
