Exceptional Handling

In Python, exception handling lets you gracefully manage runtime errors using try, except, else, and finally blocks, ensuring your program doesn’t crash unexpectedly. 

Errors vs Exceptions

Errors: Syntax issues detected before execution (e.g., SyntaxError).

Exceptions: Runtime problems (e.g., ZeroDivisionError, FileNotFoundError) that can be caught and handled.

syntax
try:
    risky_code()
except SomeException:
    handle_error()
else:
    run_if_no_error()
finally:
    cleanup_tasks()


Common build in exceptions

TypeError → Raised when an operation or function is applied to an object of inappropriate type (e.g., adding a string to an integer).

IndexError → Raised when trying to access a list index that doesn’t exist.

KeyError → Raised when a dictionary key is not found.

AttributeError → Raised when an invalid attribute reference is made.

ZeroDivisionError -> divide by zero

ImportError → Raised when an import fails.

FileNotFoundError → Raised when trying to open a file that doesn’t exist.

IOError → Raised when an input/output operation fails.

NameError → Raised when a variable is not defined.

RuntimeError → Raised when an error is detected that doesn’t fall into other categories.

OverflowError → Raised when a calculation exceeds the limits of the numeric type.

MemoryError → Raised when an operation runs out of memory.

StopIteration → Raised to signal the end of an iterator.

AssertionError → Raised when an assert statement fails.

```python
#Name error
try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
```

```python
#Finally
#The finally block, if specified, will be executed regardless if the try block raises an error or not.

try:
  print('Hello')
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
```

```python
#Raise an exception
x = -1
if x < 0:
  raise Exception("Sorry, no numbers below zero")

```

```python
x = "hello"

if not type(x) is int:
  raise TypeError("Only integers are allowed")
```

```python
#IO error:
try:
    with open("nonexistent_file.txt", "r") as f:
        data = f.read()
except IOError as e:
    print("I/O error occurred:", e)
```

```python
#runtime error:
def divide(a, b):
    if b == 0:
        raise RuntimeError("Division by zero is not allowed")
    return a / b

try:
    result = divide(10, 0)
except RuntimeError as e:
    print("RuntimeError occurred:", e)
```
