```python
website= ["www.google.com","www.amazon.in","www.yahoo.org"]
# x=website[0].split('.')
y=[]
for i in website:
    x= i.split('.')
    #-> y+=x[-1] #['c', 'o', 'm', 'i', 'n', 'o', 'r', 'g']
    y.append(x[-1]) #['com', 'in', 'org']

print(y)

#using list comprehension
y=[i.split('.')[-1] for i in website]
print(y)
```

```python
website = ["www.google.com", "www.amazon.in", "www.yahoo.org"]
y = [i.split('.')[-1] for i in website]
print(y)  # Output: ['com', 'in', 'org']
```

```python
#List comprehension in Python is a compact way to create lists by combining looping and conditional logic into a single line

squares = [x**2 for x in range(1, 6)]
print(squares)  # [1, 4, 9, 16, 25]
```

```python
even= [x for x in range(1, 11) if x%2==0]
print(even) # [2, 4, 6, 8, 10]
```

```python
#Nested Loops in Comprehension
pairs = [(x, y) for x in range(3) for y in range(2)]
print(pairs)  # [(0,0), (0,1), (1,0), (1,1), (2,0), (2,1)]
```

```python
labels = [str(x)+" is even" if x % 2 == 0 else str(x)+" is odd" for x in range(5)]
print(labels)  # ['0 is even', '1 is odd', '2 is even', '3 is odd', '4 is even']
```
