```python
import json
#If you have a JSON string, you can parse it by using the json.loads() method.
#The result will be a Python dictionary.
# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"])
print(y.keys())
```

```python

import json

with open("Python basics/data.json", "r") as f:
    data = json.load(f)

print(data)

print(data["company"])  # TechNova
print(data["year"])     # 2025
```


JSON Type	Python Equivalent
Object	    dict
Array	    list
String	    str
Number	    int/float
true/false	True/False
null	    None

```python
print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

print(json.dumps(x)) #Converts Python objects into JSON-formatted strings.
print(json.dumps(x, indent=4))
print("superator")


```

```python
print(json.dumps(x, indent=4, separators=(". ", " - "))) #Use the separators parameter to change the default separator:
```

```python
json.dumps(x, indent=4, sort_keys=True) #Use the sort_keys parameter to specify if the result should be sorted or not:
```

```python
#Writes Python objects directly into a JSON file.
person = {"name": "Arun", "age": 25}

with open("Python basics/data.json", "w") as f:
    json.dump(person, f)

with open("Python basics/data.json", "r") as f:
    data = json.load(f)

print(data)


```
