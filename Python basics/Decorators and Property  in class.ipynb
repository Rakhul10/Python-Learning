A decorator is a callable that modifies the behavior of another function or method.

In classes, decorators are often used to manage attributes or add functionality to methods.


Property is a built-in decorator that lets you define managed attributes.

Instead of writing explicit getter and setter methods, you can use @property to make attribute access look natural.

Combination: You can stack decorators to add logging, validation, or other behaviors to properties.


Traditional Get/Set                             	Property Decorator
Requires explicit get_x() and set_x() methods	        Access looks like normal attribute usage
More verbose and less Pythonic	                        Cleaner, more readable code
Harder to enforce backward compatibility	            Easier to maintain APIs without breaking code
Example: obj.get_temp()	                                Example: obj.temp

When to Use @property
Validation: Ensure values meet conditions (e.g., temperature above absolute zero).

Computed attributes: Return values derived from other data.

Read-only attributes: Expose values without allowing modification.

Backward compatibility: Change internal logic without breaking external API.

```python
def shout(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper

@shout
def greet(name):
    return f"Hello {name}"

greet("Ramesh")
```

```python
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    @property #@property alone = Read-only (getter only)
    def width(self):       # getter
        return self._width

    @width.setter
    def width(self, value):  # setter
        if value <= 0:
            raise ValueError("Width must be positive")
        self._width = value

    @property
    def height(self):      # getter
        return self._height

    @height.setter
    def height(self, value): # setter
        if value <= 0:
            raise ValueError("Height must be positive")
        self._height = value

    @property
    def area(self):        # read-only property
        return self._width * self._height
    
rect = Rectangle(10, 5)

print(rect.area)   # 50 (calls area property)
rect.width = 20    # updates width using setter
print(rect.area)   # 100 (area recalculates automatically)

```

```python
#Decorators + Property Together
#A decorator to log whenever a property is accessed.
#A property to safely manage the balance.

def log_access(func):
    def wrapper(*args, **kwargs):
        print(f"[LOG] Accessing {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

class BankAccount:
    def __init__(self, owner, balance=0):
        self._owner = owner
        self._balance = balance

    @property
    @log_access
    def balance(self):   # getter with logging
        return self._balance

    @balance.setter
    def balance(self, value):  # setter with validation
        if value < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = value

acc = BankAccount("Ramesh", 500)

print(acc.balance)   # Logs "[LOG] Accessing balance" → 500
acc.balance = 800    # Updates balance safely
print(acc.balance)   # Logs again → 800
```
