Encapsulation is about hiding data and wrapping it with methods. It protects an object's internal state from unauthorized direct modification.

Abstraction is about hiding complexity. It shows only the essential features of an object to the outside world, hiding how those features actually work under the hood.

In Python, encapsulation is achieved by restricting direct access to certain methods and attributes. This prevents accidental data corruption. We use underscores to denote the access levels of a variable or method:

Protected (_variable): A single underscore is a convention warning other developers that this attribute shouldn't be accessed outside the class (though Python won't strictly stop them).

Private (__variable): A double underscore triggers Name Mangling, where Python actively changes the internal variable name to make it much harder to access directly from outside the class.

To safely view or modify these hidden attributes, we use Getter and Setter methods.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute

    # Getter method to safely access private data
    def get_balance(self):
        return self.__balance

    # Setter method to safely modify private data with validation rules
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Invalid deposit amount!")

account = BankAccount("Anand", 1000)

#print(account.__balance)  # ❌ Throws an AttributeError (Protected by name mangling)
print(account.get_balance())  #  Output: 1000

account.deposit(500)
print(account.get_balance())  #  Output: 1500
```

In Python, abstraction is implemented using the built-in abc (Abstract Base Classes) module. You define an Abstract Class that acts as a blueprint or template, forcing child classes to implement specific methods

```python
from abc import ABC, abstractmethod

# Abstract Class (Blueprint)
class Vehicle(ABC):
    
    @abstractmethod
    def start_engine(self):
        """Every child class MUST implement this specific method"""
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started: Vroom Vroom!"

class Truck(Vehicle):
    def engine(self):
        return "Truck engine started: Rumble Rumble!"
    
# my_vehicle = Vehicle() # ❌ Throws a TypeError: Can't instantiate abstract class

car = Car()
print(car.start_engine())  # Output: Car engine started: Vroom Vroom!
truck = Truck()
print(truck.engine())  # TypeError: Can't instantiate abstract class Truck without an implementation for abstract method 'start_engine'
```
